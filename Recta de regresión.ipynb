```python
processDf=pd.read_excel("Date_Fruit_Datasets.xlsx")
processDf.head()
```

```python
fig=plt.figure(figsize=(14,14))
plt.scatter(processDf['PERIMETER'],processDf['AREA'])
plt.plot(processDf['PERIMETER'],processDf['AREA'])
plt.xlabel('PERIMETER')
plt.ylabel('AREA')
plt.grid()
```

```python
processDf.keys()
```

```python
nPERIMETER=processDf['PERIMETER'].values.reshape(-1,1)
nAREA=processDf['AREA'].values.reshape(-1,1)
linear_regresor=LinearRegression()
linear_regresor.fit(nPERIMETER,nAREA)
nAREA_pred=linear_regresor.predict(nPERIMETER)
nAREA_pred
```

```python
#Ecuacion de la linea y=mx+c
m=linear_regresor.coef_[0][0]
c=linear_regresor.intercept_[0]
label=r"$nAREA=%0.35f+nPERIMETER%+0.35f$"%(m,c)
print(label)
```

```python
fig=plt.figure(figsize=(14,14))
plt.scatter(processDf['PERIMETER'],processDf['AREA'])
plt.plot(processDf['PERIMETER'],processDf['AREA'], label='Measured nAREA')
plt.plot(nPERIMETER,nAREA_pred, color='red', label=label)
plt.xlabel('PERIMETER')
plt.ylabel('AREA')
plt.legend()
plt.grid()
```
